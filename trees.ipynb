```python
#bst
class bst:
    def __init__(self,key):
        self.key=key
        self.l=None
        self.r=None
#insertion
    def insert(self,data):
        if (self.key==None):
            self.key=data
            print("empty")
            return
        if(self.key==data):# for removing the dupicate values
            return
        if(self.key>data):
            if(self.l):
                self.l.insert(data)
            else:
                self.l=bst(data)
        else:
            if(self.r):
                self.r.insert(data)
            else:
                self.r=bst(data)
#traversal algorithm
    #preorder ROOT,LEFT,RIGHT
    def preorder(self):
        print(self.key,end =' ')
        if self.l:
            self.l.preorder()
        if self.r:
            self.r.preorder()
    #inorder LEFT,ROOT,RIGTH
    def inorder(self):
        if self.l:
            self.l.inorder()
        print(self.key,end =' ')
        if self.r:
            self.r.inorder()
    #postorder LEFT,RIGHT,ROOT
    def postorder(self):
        if self.l:
            self.l.postorder()
        if self.r:
            self.r.postorder()
        print(self.key,end =' ')
#search method 
    def search(self,data):
        if self.key==data:
            print("node is found in the tree")
            return
        if(self.key>data):
            if(self.l):
                self.l.search(data)
            else:
                print("node is not present in tree")
        else:
            if(self.r):
                self.r.search(data)
            else:
                print("node is not present in tree")
#max and min
    def min(self):
        current=self
        while current.l:
            current=current.l
        print("MIN element in tree is::",current.key)
    def max(self):
        current=self
        while current.r:
            current=current.r
        print("MAX element in tree is::",current.key)

#deletion method
    def delete(self,data):
        if (self.key==None):
            print("empty")
            return
        # deletion of root node
        if (self.l==None and self.r==None):
            if(self.l==self):
                self.l=None
            else:
                self.r=None
        if(data<self.key):
            if(self.l):
                self.l=self.l.delete(data)
            else:
                print("given node is not present in the tree")
        elif(data>self.key):
            if(self.r):
                self.r=self.r.delete(data)
            else:
                print("given node is not present in the tree")
        else:
            if(self.l==None):
                temp=self.r
                self=None
                return temp
            if(self.r==None):
                temp=self.l
                self=None
                return temp
            node=self.r
            while node.l:
                node=node.l
            self.key=node.key
            self.r=self.r.delete(node.key)
        return self

root=bst(10)
m=[6,3,1,6,98,3,7,13,22,2.5,13.56]
print("inserting elments :: ",end=" ")
for i in m:
    root.insert(i)
    print(i,end=" ")
print()
root.search(6)
print()
root.min()
root.max()
print()
print("preorder :-ROOT,LEFT,RIGHT")
root.preorder()
print()
print("inorder  :-LEFT,ROOT,RIGTH")
root.inorder()
print()
print("postorder:-LEFT,RIGHT,ROOT")
root.postorder()
print()
root.delete(98)
print()
print("after deleting:-")
root.preorder()
```
